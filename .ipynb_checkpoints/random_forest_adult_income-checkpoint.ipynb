```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier





# create the csv file with headers from the data provided from UCI machine learning repository

df = pd.read_csv("adult_data.txt", header=None, names=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income'], na_values=[' ?'])


df


# preprocess the data

# workclass, occupation, and native-country all have '?' values. May have to delete their records

# df[df['workclass'] == ' ?']

# df['workclass'] = df['workclass'].str.lstrip()
# get rid of rows that show '?'
# df = df[df['workclass'] != ' ?']
```

```python
# Make a heatmap to determine which fields correlate most to target income

fig, (ax) = plt.subplots(1,1,figsize=(10,6))

hm = sns.heatmap(df.corr(),
                ax=ax,
                cmap="icefire",
                annot=True,
                fmt='.2f',
                linewidths=.05)

fig.subplots_adjust(top=0.92)
fig.suptitle('US DATA INCOME Correlation Heatmap',
                fontsize=14,
                fontweight='bold')
```

```python
# Drop education column since it's just the categorical equivalent to education-num

del df['education']
```

```python
# NaN values in our dataset
df.isnull().sum()
```

```python
df.income.value_counts()
```

```python
'''Encoding - We need to encode categorical values to numerical ones. The missing (NaN) values will
take on the mean of each column
'''

one_hot_encoding = pd.get_dummies(df)
one_hot_encoding.fillna(one_hot_encoding.mean(), inplace=True)
one_hot_encoding["income"] = one_hot_encoding['income_ >50K']
del one_hot_encoding['income_ >50K']
del one_hot_encoding['income_ <=50K']
sns.set(font_scale=2.8)
sns.clustermap(one_hot_encoding.corr(), metric='correlation',figsize=(50,50))
```

```python
# split dataset into train and test
x = one_hot_encoding.iloc[:,0:-1]
y = one_hot_encoding.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=42)
```

```python
rf = RandomForestClassifier(max_depth=25,n_estimators=200, min_samples_leaf=1, min_samples_split=50, criterion='entropy', oob_score=True, random_state=42)
rf.fit(X_train, y_train)
```

```python
predicted = rf.predict(X_test)
accuracy = accuracy_score(y_test,predicted)
print(f'Out-of-bag score estimate: {rf.oob_score_:.3}')
print(f'Mean accuracy score: {accuracy:.3}')
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, predicted)
fig, (ax) = plt.subplots(1,1, figsize=(8,7))
sns.heatmap(cm, annot=True, ax=ax, fmt='g')
# labels, titles, ticks
ax.set_xlabel('Predicted Labels')
ax.set_ylabel('Actual Labels')
ax.set_title('Confusion Matrix')
sns.set(font_scale=1.4)
ax.xaxis.set_ticklabels(['income > 50K', 'income < 50K'])
ax.yaxis.set_ticklabels(['income > 50K', 'income < 50K'])
```
